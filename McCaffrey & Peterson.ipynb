```python
import pandas as pd
```

<h2>Christian McCaffrey 2023

```python
url_cmac = "https://www.pro-football-reference.com/players/M/McCaCh01.htm"
cmac = pd.read_html(url_cmac)
cmac = cmac[1]
display(cmac)
```

```python
cmac.columns = cmac.columns.droplevel(0)
display(cmac)
```

```python
cmac_2023 = cmac.query('Year == "2023" ')
display(cmac_2023)
```

```python
cmac_2023 = cmac_2023[["Year", "G", "Yds", "TD"]]
display(cmac_2023)
```

```python
cmac_2023.columns = ["Year", "G", "Rush_Yds", "Rec_Yds", "Rush_TDs", "Rec_TDs"]
display(cmac_2023)
```

```python
cmac_2023.dtypes
```

```python
cmac_2023["Total_Yds"] = cmac_2023["Rush_Yds"] + cmac_2023["Rec_Yds"]
cmac_2023["Total_TDs"] = cmac_2023["Rush_TDs"] + cmac_2023["Rec_TDs"]
cmac_2023["Player"] = "Christian McCaffrey"
display(cmac_2023)
```

```python
cmac_2023_final = cmac_2023[["Year", "Player", "G", "Rush_Yds", "Rec_Yds", "Total_Yds", "Rush_TDs", "Rec_TDs",  "Total_TDs"]]
display(cmac_2023_final)
```
